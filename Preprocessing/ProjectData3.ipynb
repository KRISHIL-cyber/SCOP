```python
def create_csv(filename, output_file):
  print(filename)
  print()
  df=pd.read_csv(filename)
  df.tail()
  df=df.drop(df.columns[0], axis=1)
  df1= pd.concat([df,labels], axis=1)
  df1 = df1[df1['class'] != 'x']
  df1.info()
  df1.to_csv(output_file, index=False)
```
